# importing libraries for data visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
cars_ds = pd.read_csv("bar_chart_data.csv")
cars_ds = cars_ds.sort_values("Cars Listings")
```

# Bar Graphs

```python
plt.figure(figsize=(9,6))
plt.bar(x = cars_ds.Brand,height=cars_ds["Cars Listings"],color = "rgbwymc")
plt.title("Cars Listing By Brands",fontsize = 16,fontweight = "bold")
plt.ylabel("Number of listing",fontsize=20)
plt.xticks(rotation = 45,fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

# PIE CHART

```python
plt.pie(cars_ds["Cars Listings"],labels=cars_ds["Brand"],radius=3,textprops={"size":"x-large","fontweight":"bold"})
plt.show()
```

```python
fuel_engine_df = pd.read_csv("pie_chart_data.csv")
```

```python
plt.figure(figsize=(10,10))
plt.pie(fuel_engine_df["Number of Cars"],
        labels=fuel_engine_df["Engine Fuel Type"].values,
        autopct="%.2f%%",
        textprops={"size":"x-large","fontweight":"bold","rotation":"30"},
       colors=sns.color_palette("colorblind"))
plt.legend()
plt.title("Cars by engine fule type",fontsize = 18,fontweight="bold")
plt.show()
```

# STACKED AREA PLOT

```python
car_sales_df = pd.read_csv("stacked_area_chart_data.csv")
```

```python
colors = ["#011638","#7e2987","ef2026","yellow"]
labels = ["Gas","Petrol","Diesel","Other"]
plt.figure(figsize=(10,6))
plt.stackplot(car_sales_df["Year"],
              car_sales_df["Gas"],
              car_sales_df["Petrol"],
              car_sales_df["Diesel"],
              car_sales_df["Other"],
              colors=colors,
              edgecolor = 'none')
plt.legend(labels = labels,fontsize = 13, loc = "upper left")
plt.ylabel("No of Cars",fontsize = 13)
plt.xlabel("Years",fontsize = 13)
plt.title("Popularity of type of cars from 1982 to 2016",fontsize = 15,fontweight = "bold")
plt.xticks(car_sales_df["Year"],rotation = 45)
plt.show()
```

# Line Chart

```python
stock_ds = pd.read_csv("line_chart_data.csv")
```

```python
stock_ds["New_Date"] = pd.to_datetime(stock_ds["Date"])
```

```python
stock_ds = stock_ds.drop("Date",axis=1)
```

```python
new_order = ["New_Date","GSPC500","FTSE100"]
stock_ds = stock_ds[new_order]
```

```python
stock_ds = stock_ds.sort_values("New_Date")
```

```python
labels = ["S&P 500","FTSE 100"]
plt.figure(figsize=(20,10))
plt.plot(stock_ds["New_Date"],stock_ds["GSPC500"])
plt.plot(stock_ds["New_Date"],stock_ds["FTSE100"])
plt.xlabel("Year")
plt.ylabel("Return")
plt.legend(labels,fontsize="large")
plt.title("Returns in S&P and FTSE",fontsize = 20,fontweight = "bold")
plt.show()
```

```python
stock_ds_new = stock_ds[(stock_ds.New_Date >= '2008-07-01') & (stock_ds.New_Date <= '2008-12-31')]
```

```python
labels = ["S&P 500","FTSE 100"]
plt.figure(figsize=(20,10))
plt.plot(stock_ds_new["New_Date"],stock_ds_new["GSPC500"])
plt.plot(stock_ds_new["New_Date"],stock_ds_new["FTSE100"])
plt.xlabel("Year")
plt.ylabel("Return")
plt.legend(labels,fontsize="large")
plt.title("Returns in S&P and FTSE",fontsize = 20,fontweight = "bold")
plt.show()
```

```python
real_estate_df = pd.read_csv("histogram_data.csv")
```

# HISTOGRAM

```python
plt.figure(figsize=(10,10))
plt.hist(real_estate_df["Price"],bins=10,color="#108A99")
plt.show()
```

```python
real_estate_df.Price = np.log(real_estate_df.Price)
plt.figure(figsize=(10,10))
plt.hist(real_estate_df["Price"],bins=10,color="#108A99")
plt.title("Propert price distribution",fontweight = "bold")
plt.xlabel("log prices in $",fontweight = "bold")
plt.ylabel("Number of price",fontweight = "bold")
```

# Scatter Plot

```python
real_estat_df_Ca = pd.read_csv("scatter_data.csv")
```

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(real_estat_df_Ca["Area (ft.)"],real_estat_df_Ca["Price"],
            alpha=0.8,c=real_estat_df_Ca["Building Type"],cmap="mako")
plt.legend(*scatter.legend_elements(),
          loc="upper left",
          title = "Building Type")
plt.title("Relationship between area and price of california real estate")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(real_estat_df_Ca["Area (ft.)"],
                real_estat_df_Ca["Price"],
               hue=real_estat_df_Ca["Building Type"],
               s=100)
plt.show()
```

```python
from matplotlib.ticker import PercentFormatter
barandline = pd.read_csv("bar_line_chart_data.csv")
```

```python
fig, ax = plt.subplots(figsize = (10,7))
ax.bar(barandline["Year"],
       barandline["Participants"],
       color = "k")
ax.set_ylabel("Number of participants",
             weight = "bold")
ax.tick_params(axis = "y",width = 2,
              labelsize = "large")
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=0.1))
ax1.plot(barandline["Year"],
         barandline["Python Users"],
         color = "blue",
         marker = "*")
ax1.set_ylabel("Python User",
               weight = "bold",
               color = "blue")
ax1.tick_params(axis = "y",
                width = 2,
                labelsize = "large")
ax.set_title("Survey from 2012-2019",fontsize=20,weight="bold")
plt.show()
```
